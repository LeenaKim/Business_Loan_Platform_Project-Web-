## 함수

- 한글찾는 함수

```python

def find_korean(data_list):
    result = []
    for data in data_list:
        result += re.compile('[|가-힣]+').findall(data)
    return result
```

 - 한글 값을 지우는 함수

```python
def remove_korean(data_list):
    for i in range(len(data_list)):
        data_list.iloc[i] = re.sub('[|가-힣]+', '999', str(data_list.iloc[i]))
    
```

 - 한글이 들어있을시 전체 값을 '999'로 

```python
def remove_colval_kor(data_list):
    for i in range(len(data_list)):
       if (re.search('[|가-힣]+', str(data_list.iloc[i])) is not None):
          data_list.iloc[i] = '999'
        
```

- '-' 포함시 전체 값을 '999'로 

```python
def remove_colval_hypen(data_list):
    for i in range(len(data_list)):
       if (re.search('-', str(data_list.iloc[i])) is not None):
          data_list.iloc[i] = '999'
        
```

- '-' 지우는 함수

```python
def remove_hypen(data_list):
    for i in range(len(data_list)):
        data_list.iloc[i] = re.sub('-', '', str(data_list.iloc[i]))
```

- 대문자를 소문자로 바꿔주는 함수

```python
def upper_to_lower(data_list):
    for i in range(len(data_list)):
        data_list.iloc[i] = data_list.iloc[i].lower()
```

- 'http://www' 지우는 함수

```python
def remove_http(data_list):
    for i in range(len(data_list)):
        data_list.iloc[i] = re.sub('http://', '', str(data_list.iloc[i]))
        data_list.iloc[i] = re.sub('www.', '', str(data_list.iloc[i]))
```

- '(' 지우는 함수 

```python
def remove_brace(data_list):
    for i in range(len(data_list)):
        data_list.iloc[i] = re.sub('[()]', '', str(data_list.iloc[i]))
```

- 공백 지우는 함수

```python
def remove_space(data_list):
    for i in range(len(data_list)):
        data_list.iloc[i] = re.sub('[\s]', '', str(data_list.iloc[i]))
```

- '+' 지우는 함수

```python
def remove_plus(data_list):
    for i in range(len(data_list)):
        data_list.iloc[i] = re.sub('[+]', '', str(data_list.iloc[i]))
```

```python
# 반복문을 돌며 복수개의 대표 이름을 각각의 행으로 담을 새로운 df 생성
new_ceo_df = pd.DataFrame(columns=['corp_code', 'ceo_nm'])
```

```python
len(new_ceo_df)
```

```python
# 대표명에 쉼표가 들어갔을시 쉼표로 split한 리스트를 새로운 df에 저장
cnt = 0
for i in range(len(ceo_df2)):
    if(re.search('[|가-힣]+', str(ceo_df2['ceo_nm'].iloc[i])) is not None):
        if("," in ceo_df2['ceo_nm'].iloc[i] ):
            names = ceo_df2['ceo_nm'].iloc[i].split(',')
            for j in range(len(names)):
                new_ceo_df.loc[cnt] = [ceo_df2['corp_code'].iloc[i], names[j]]
                cnt = cnt + 1
        elif "." in ceo_df2['ceo_nm'].iloc[i]:
            names = ceo_df2['ceo_nm'].iloc[i].split('.')
            for j in range(len(names)):
                new_ceo_df.loc[cnt] = [ceo_df2['corp_code'].iloc[i], names[j]]
                cnt = cnt + 1
        else:
            new_ceo_df.loc[cnt] = [ceo_df2['corp_code'].iloc[i], ceo_df2['ceo_nm'].iloc[i]]
    else:
        new_ceo_df.loc[cnt] = [ceo_df2['corp_code'].iloc[i], ceo_df2['ceo_nm'].iloc[i]]
    cnt = cnt + 1
    
    
```
